````python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import chart_studio.plotly as py
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot

%matplotlib inline
init_notebook_mode(connected=True)
````

### 1. Load and Visualize Data

Let's start by loading the entire dataset.

````python
sales_df = pd.read_csv('/kaggle/input/store-demand-forecasting-dataset-nndl/train.csv', parse_dates=['date'])
````

````python
#Visualizing the time-series sales records as a sequence plot

daily_trace = go.Scatter(x=sales_df['date'], y=sales_df['sales'])
layout = go.Layout(title='Daily sales', xaxis=dict(title='Date'), yaxis=dict(title='Sales'))
fig = go.Figure(data=[daily_trace], layout=layout)
iplot(fig)
````

### 2. Computing Some Statistics from the  Data

````python
df_numeric = sales_df.drop(columns='date')
````

````python
df_numeric.describe()
````

````python
print('Min date from train set: %s' % sales_df['date'].min().date())
print('Max date from train set: %s' % sales_df['date'].max().date())
````

````python
sales_by_store = sales_df.groupby(['store', 'date'], as_index=False)['sales'].sum()
sales_by_item = sales_df.groupby(['item', 'date'], as_index=False)['sales'].sum()
````

````python


#Per-store daily sales visualization

store_traces = []
for store in sales_by_store['store'].unique():
    curr_store_data = sales_by_store[(sales_by_store['store'] == store)]
    store_traces.append(go.Scatter(x=curr_store_data['date'], y=curr_store_data['sales'], name=f'Store {store}'))

layout = go.Layout(title='Store daily sales', xaxis=dict(title='Date'), yaxis=dict(title='Sales'))
fig = go.Figure(data=store_traces, layout=layout)
iplot(fig)

````

````python
#Per-item daily sales visualization

item_traces = []
for item in sales_by_item['item'].unique():
    curr_item_data = sales_by_item[(sales_by_item['item'] == item)]
    item_traces.append(go.Scatter(x=curr_item_data['date'], y=curr_item_data['sales'], name=f'Item {item}'))

layout = go.Layout(title='Item daily sales', xaxis=dict(title='Date'), yaxis=dict(title='Sales'))
fig = go.Figure(data=item_traces, layout=layout)
iplot(fig)
````

1. Our dataset is balanced. The sales data for each store is distributed approximately equally in our dataset.
2. From our dataset we can infer that the sales volume increases year by year across all the different stores and across all the different items. Sales volume peaks in a particular month every year. This can be a result of holidays or any such festivities which needs to be investigated further
